```python
import cv2
import os

# Function to extract frames and save as images
def extract_frames(video_path, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Open the video file
    cap = cv2.VideoCapture(video_path)
    
    # Check if the video file is opened successfully
    if not cap.isOpened():
        print("Error: Couldn't open the video file.")
        return
    
    frame_count = 0
    
    while True:
        # Read frame-by-frame
        ret, frame = cap.read()
        
        if not ret:
            break  # Break the loop when no more frames are available
        
        # Save the current frame as an image
        frame_filename = os.path.join(output_folder, f"frame_{frame_count:04d}.jpg")
        cv2.imwrite(frame_filename, frame)
        
        # Print progress
        print(f"Saving frame {frame_count}")
        
        frame_count += 1
    
    # Release the video capture object
    cap.release()
    print(f"Frames saved to {output_folder}")

# Example usage:
video_path = '/Users/annastuckert/Documents/GitHub/DeepEnsampleGUI/napari-threshold/Circle_12_12_43_17.avi'  # Path to your video file
output_folder = 'frames'  # Folder where the frames will be saved

extract_frames(video_path, output_folder)
```
